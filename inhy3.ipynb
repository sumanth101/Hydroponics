```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/env.csv')
df.head()
```

```python
# Initialize dictionaries to store the sum and count of pH and EC values for each bay number
temp_sum = {}
count = {}

# Iterate over each row in the data
for index, row in df.iterrows():
    bay = row["bay"]
    temp = row["temp_c"]  # Fixed variable name

    # Update the sum and count of pH and EC values for the bay number
    temp_sum[bay] = temp_sum.get(bay, 0) + temp
    count[bay] = count.get(bay, 0) + 1

# Calculate the average pH and EC values for each bay number
temp_avg = {bay: temp_sum[bay] / count[bay] for bay in temp_sum}

# Print the average pH and EC values for each bay number
for bay in range(201, 211):
    print(f"Bay {bay}: Average temp = {temp_avg.get(bay, 0)}")
```

```python
# Initialize dictionaries to store the sum and count of pH and EC values for each bay number
hum_sum = {}
count = {}

# Iterate over each row in the data
for index, row in df.iterrows():
    bay = row["bay"]
    hum = row["relative_humidity_%"]  # Fixed variable name

    # Update the sum and count of pH and EC values for the bay number
    hum_sum[bay] = hum_sum.get(bay, 0) + hum
    count[bay] = count.get(bay, 0) + 1

# Calculate the average pH and EC values for each bay number
hum_avg = {bay: hum_sum[bay] / count[bay] for bay in hum_sum}

# Print the average pH and EC values for each bay number
for bay in range(201, 211):
    print(f"Bay {bay}: Average Humidity = {hum_avg.get(bay, 0)}")
```

```python
data=pd.read_csv('/content/env2.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Extract the temp, humidity, and nitrogen percent values from the dataset
temp_values = data["temp"].values
hum_values = data["humidity"].values
nitrogen_percent = data["N"].values

# Create a 3D plot
fig = plt.figure()
ax = plt.axes(projection="3d")

# Plot the data
ax.scatter3D(temp_values, hum_values, nitrogen_percent, c=nitrogen_percent, cmap='viridis')

# Set x, y, and z labels
ax.set_xlabel("Temperature")
ax.set_ylabel("Humidity")
ax.set_zlabel("Nitrogen %")

# Add a colorbar
cbar = plt.colorbar(ax.collections[0])
cbar.set_label("Nitrogen %")

# Show the plot
plt.show()

# Find the index of the maximum nitrogen percent value
max_index = np.argmax(nitrogen_percent)

# Get the corresponding temp and humidity values
max_temp = temp_values[max_index]
max_hum = hum_values[max_index]


# Print the result
print("Point with highest nitrogen percentage:")
print("temperature:", max_temp)
print("Humidity:", max_hum)
```
