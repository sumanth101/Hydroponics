```python
# Read the data from the CSV file
data = pd.read_csv("/content/bay,ec,pH.csv")
data.head()
```

```python
data.info()
```

```python
# Initialize dictionaries to store the sum and count of pH and EC values for each bay number
pH_sum = {}
ec_sum = {}
count = {}

# Iterate over each row in the data
for index, row in data.iterrows():
    bay = row["bay"]
    pH = row["pH"]
    ec = row["EC"]

    # Update the sum and count of pH and EC values for the bay number
    pH_sum[bay] = pH_sum.get(bay, 0) + pH
    ec_sum[bay] = ec_sum.get(bay, 0) + ec
    count[bay] = count.get(bay, 0) + 1

# Calculate the average pH and EC values for each bay number
pH_avg = {bay: pH_sum[bay] / count[bay] for bay in pH_sum}
ec_avg = {bay: ec_sum[bay] / count[bay] for bay in ec_sum}

# Print the average pH and EC values for each bay number
for bay in range(201, 211):
    print(f"Bay {bay}: Average pH = {pH_avg.get(bay, 0)}, Average EC = {ec_avg.get(bay, 0)}")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Read the data from the CSV file
data = pd.read_csv("/content/1.csv")

# Extract the pH, EC, and nitrogen percent values from the dataset
pH_values = data["pH"].values
ec_values = data["EC"].values
nitrogen_percent = data["N"].values

# Create a 3D plot
fig = plt.figure()
ax = plt.axes(projection="3d")

# Plot the data
ax.scatter3D(pH_values, ec_values, nitrogen_percent, c=nitrogen_percent, cmap='viridis')

# Set x, y, and z labels
ax.set_xlabel("pH")
ax.set_ylabel("EC")
ax.set_zlabel("Nitrogen %")

# Add a colorbar
cbar = plt.colorbar(ax.collections[0])
cbar.set_label("Nitrogen %")

# Show the plot
plt.show()
```

```python
# Find the index of the maximum nitrogen percent value
max_index = np.argmax(nitrogen_percent)

# Get the corresponding EC and pH values
max_ec = ec_values[max_index]
max_ph = pH_values[max_index]

# Print the result
print("Point with highest nitrogen percentage:")
print("EC:", max_ec)
print("pH:", max_ph)
```
